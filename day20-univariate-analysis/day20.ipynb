b. Distplot

A `KDE` (Kernel Density Estimate) plot is a smooth, continuous approximation of a histogram. It estimates the probability density function (PDF) of a continuous variable, making it easier to see the underlying distribution of the data.

Distplot : it also tells about the skewness of the data

```python

sns.displot(df, x='Age', kind='kde') # For KDE only
sns.displot(df, x='Age', kde=True)  # For histogram with KDE
```
