```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package (In LPA)')
plt.show()
```

Splitting X & y

```python
X = df.iloc[:, 0:1]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
X_test.shape
```

```python
X_test
```

```python
y_test
```

```python
y_pred = lr.predict(X_test)
y_pred[0]
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train, lr.predict(X_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package (in lpa)')
```

```python
# slope
m = lr.coef_
m
```

```python
# intercept
b = lr.intercept_
b
```
