```python
df['experience'] = df['experience'].apply(lambda x: w2n.word_to_num(x) if isinstance(x, str) else x)
```

```python
df
```

```python
df['experience'].median()
```

```python
df['experience'] = df['experience'].fillna(df['experience'].median())
```

```python
df['test_score(out of 10)'].median()
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())
```

```python
df.isnull().sum()
```

```python
X, y = df.drop('salary($)', axis=1), df['salary($)']
```

```python
X
```

```python
y
```

Model - LR

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, y)
```

```python
# [w1, w2, w3]
lr.coef_
```

```python
# w0
lr.intercept_
```

```python
# y = w0 + w1*x1 + w2*x2 + w3*x3
# salary = w0 + w1*experience + w2*test_score + w3*interview_score
```

''' Predict Salaries of candidates who has
1. 2 yr experience, 9 test score, 6 interview score
2. 12 yr experience, 10 test score, 10 interview score
'''

```python
lr.predict([[2, 9, 6]])
```
