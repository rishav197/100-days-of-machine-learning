```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data = pd.read_csv('/content/drive/MyDrive/day72-Hierarchical-clustering/hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
customer_data.head()
```

```python
customer_data.columns
```

```python
data = customer_data[['Annual Income (k$)', 'Spending Score (1-100)']]
data
```

Plot for Dendogram

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

### Applying Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
```

```python
labels = ac.fit_predict(data)
labels
```

```python
data.columns
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=labels, cmap='rainbow')
plt.title("Agglomerative Clustering Results")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```
