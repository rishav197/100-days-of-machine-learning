### Outliers Detection

```python
# Finding the outliers
df[(df['cgpa'] > 8.80) | (df['cgpa'] < 5.11)]
```

## Outliers Treatment

### Approach1: Trimming

```python
# Trimming

new_df = df[(df['cgpa'] < 8.80) & (df['cgpa'] > 5.11)]
new_df
```

### Approach2 : Using Z-score

```python
# Calculating the Zscore

df['cgpa_zscore'] = (df['cgpa'] - df['cgpa'].mean())/df['cgpa'].std()
```

```python
df.head()
```

```python
df[df['cgpa_zscore'] > 3]
```

```python
df[df['cgpa_zscore'] < -3]
```

```python
df[(df['cgpa_zscore'] > 3) | (df['cgpa_zscore'] < -3)]
```

```python
# Trimming
new_df = df[(df['cgpa_zscore'] < 3) & (df['cgpa_zscore'] > -3)]
```

```python
new_df
```

## Capping

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
print(lower_limit)
print(upper_limit)
```

```python
df['cgpa'] = np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
    )
)
```
