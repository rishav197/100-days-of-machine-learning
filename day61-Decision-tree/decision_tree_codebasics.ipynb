```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/day61-Decision-tree/data/salaries.csv")
df
```

```python
df.columns
```

```python
X = df.drop('salary_more_then_100k', axis='columns')
y = df['salary_more_then_100k']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
X['company_num'] = le_company.fit_transform(X['company'])
X['job_num'] = le_job.fit_transform(X['job'])
X['degree_num'] = le_degree.fit_transform(X['degree'])
```

```python
X.head(10)
```

```python
X.columns
```

```python
X_num = X.drop(['company', 'job', 'degree'], axis='columns')
X_num
```

Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_num, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

### Applying Decision Tree

```python
from sklearn import tree
```

```python
dt = tree.DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
```

```python
dt.score(X_test, y_test)
```

Prediction

```python
"""Is a person who is working in google as a sales executive and has done masters getting salary more than 100k$ """

dt.predict([[2, 2, 1]])

#salay won't be more than 100k dollar
```

```python
"""Is a person who is working in google as a buisness manager and has done bachelors getting salary more than 100k$ """
dt.predict([[2, 0, 1]])

#salay would be more than 100k dollar
```
